```python
import pandas as pd
import numpy as np
import data_work

data=pd.read_csv("ciciot2023_processed.csv")
data
```

```python
data_work.count_label_data(data["label"])
```

```python
from sklearn.model_selection import train_test_split 
import torch
traindata,test_data=train_test_split(data,test_size=0.4,random_state=42,stratify=data['label'])
data = traindata.reset_index(drop=True)
test_data = test_data.sample(frac=1).reset_index(drop=True)
client_data,server_data=train_test_split(traindata,test_size=0.03,random_state=22520540,stratify=data['label'])
client_data = client_data.sample(frac=1).reset_index(drop=True)
server_data = server_data.sample(frac=1).reset_index(drop=True)
print(f"server data quantity {len(server_data)}")
trainloaders,testloaders=data_work.data_split(client_data,"label", 16,0.01,False)
```

```python
server_data.to_csv("./dataset/server_data.csv")
test_data.to_csv("./dataset/test_data.csv")
for i in range(16):
    trainloaders[i] = trainloaders[i].sample(frac=1).reset_index(drop=True)
    trainloaders[i].to_csv(f"./dataset/trainloader_{i}.csv")
    testloaders[i]= testloaders[i].sample(frac=1).reset_index(drop=True)
    testloaders[i].to_csv(f"./dataset/testloader_{i}.csv")
```
